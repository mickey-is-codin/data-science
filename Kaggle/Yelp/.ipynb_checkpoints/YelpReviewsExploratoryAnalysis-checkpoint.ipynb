```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.set_context("paper")

import random

from wordcloud import WordCloud, STOPWORDS
from PIL import Image
```

```python
import_rows = 200000

df = pd.read_csv(
    'yelp_review.csv',
    nrows=import_rows,
)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['word_count'] = df['text'].str.count(' ') + 1
```

```python
sns.countplot(df['stars']).set_title('Review Breakdown by Stars')
total_reviews = import_rows

one_star = df[df.stars == 1].reset_index()
one_star_count = one_star.count()['stars']

two_star = df[df['stars'] == 2].reset_index()
two_star_count = two_star.count()['stars']

three_star =  df[df['stars'] == 3].reset_index()
three_star_count = three_star.count()['stars']

four_star = df[df['stars'] == 4].reset_index()
four_star_count = four_star.count()['stars']

five_star = df[df['stars'] == 5].reset_index()
five_star_count = five_star.count()['stars']

print(' \
    \nOf the %d total reviews: \
    \n%d reviews had 1 star. \
    \n%d reviews had 2 stars.\
    \n%d reviews had 3 stars.\
    \n%d reviews had 4 stars.\
    \n%d reviews had 5 stars.\n' \
    % (total_reviews, one_star_count, two_star_count, three_star_count, four_star_count, five_star_count))
```

```python
print('The median amount of words in a one star review was %d' % (one_star['word_count'].median()))
print('The median amount of words in a two star review was %d' % (two_star['word_count'].median()))
print('The median amount of words in a three star review was %d' % (three_star['word_count'].median()))
print('The median amount of words in a four star review was %d' % (four_star['word_count'].median()))
print('The median amount of words in a five star review was %d' % (five_star['word_count'].median()))

print('\nThe mean amount of words in a one star review was %d' % (one_star['word_count'].mean()))
print('The mean amount of words in a two star review was %d' % (two_star['word_count'].mean()))
print('The mean amount of words in a three star review was %d' % (three_star['word_count'].mean()))
print('The mean amount of words in a four star review was %d' % (four_star['word_count'].mean()))
print('The mean amount of words in a five star review was %d' % (five_star['word_count'].mean()))

print('\nThe largest amount of words in a one star review was %d' % (one_star['word_count'].max()))
print('The largest amount of words in a two star review was %d' % (two_star['word_count'].max()))
print('The largest amount of words in a three star review was %d' % (three_star['word_count'].max()))
print('The largest amount of words in a four star review was %d' % (four_star['word_count'].max()))
print('The largest amount of words in a five star review was %d' % (five_star['word_count'].max()))

print('\nThe standard deviation of words in a one star review was %d' % (one_star['word_count'].std()))
print('The standard deviation of words in a two star review was %d' % (two_star['word_count'].std()))
print('The standard deviation of words in a three star review was %d' % (three_star['word_count'].std()))
print('The standard deviation of words in a four star review was %d' % (four_star['word_count'].std()))
print('The standard deviation of words in a five star review was %d' % (five_star['word_count'].std()))

long_one_star = one_star[one_star['word_count'] > 100]
long_two_star = two_star[two_star['word_count'] > 100]
long_three_star = three_star[three_star['word_count'] > 100]
long_four_star = four_star[four_star['word_count'] > 100]
long_five_star = five_star[five_star['word_count'] > 100]

print('\nThe amount of one star reviews with over 100 words was %d' % (len(long_one_star)))
print('The amount of two star reviews with over 100 words was %d' % (len(long_two_star)))
print('The amount of three star reviews with over 100 words was %d' % (len(long_three_star)))
print('The amount of four star reviews with over 100 words was %d' % (len(long_four_star)))
print('The amount of five star reviews with over 100 words was %d' % (len(long_five_star)))
```

```python
fig, ax = plt.subplots(figsize=(10,10))

overall_plot = sns.kdeplot(df['word_count']).set_title('Yelp Review Word Count by Stars Given')
one_star_plot = sns.kdeplot(one_star['word_count'], ax=ax)
two_star_plot = sns.kdeplot(two_star['word_count'], ax=ax)
three_star_plot = sns.kdeplot(three_star['word_count'], ax=ax)
four_star_plot = sns.kdeplot(four_star['word_count'], ax=ax)
five_star_plot = sns.kdeplot(five_star['word_count'], ax=ax)

plt.legend(['Overall','1 Star','2 Star','3 Star','4 Star','5 Star'])
```
